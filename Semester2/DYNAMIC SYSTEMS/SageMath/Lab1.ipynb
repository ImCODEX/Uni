```python
n = var('n')
s1(n) = 0.1*n 
s1_list=[s1(n) for n in [1..10]]
f(t,s) = 1 - s*cos(4*t)*cos(t)/(1-s^2*cos(4*t)^2*sin(t)^2)^(1/2)
x_list = [f(t-pi/2,i) for i in s1_list]
y_list = [f(t,i) for i in s1_list]
plot((x_list,y_list),(t,0,2*pi))
```
