```python
x=var('x') #EX1-a
y=function('y')(x)
eqd=x*diff(y,x,2)-diff(y,x)==x^2+x
sol1a=desolve(eqd,y)
sol1a
```

```python
sol1b=desolve(eqd,y,ics=[1,1,5]) #EX1-b
sol1b
```

```python
t=var('t') #EX3-a
x=function('x')(t)
y=function('y')(t)
deq1=diff(x,t)==-y
deq2=diff(y,t)==8*x-6*y
syst=[deq1,deq2]
sol3a=desolve_system(syst, [x,y])
sol3a
```

```python
vars=[x,y] #EX3-b
in_cond=[0,0,-3]
sol3b=desolve_system(syst, vars, in_cond)
sol3b
```

```python
t = var('t') #EX2-a
x = function('x')(t)
eqd = 4 * diff(x, t) - x == 5*sin(t)
sol2a = desolve(eqd, x)
sol2a
```
