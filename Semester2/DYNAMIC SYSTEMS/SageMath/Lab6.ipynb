```sage
t=var('t')
k=var('k')
x0=var('x0')
x=function('x')(t)
eqd=diff(x,t)==-k*x
sol(x0,k)=desolve(eqd,[x,t],ics=[0,x0])
sol
```

```sage
plot(sol(3,-1),x,-2,2)
```

```sage
t=var('t')
k=var('k')
T=function('T')(t)
Tu=var('Tu')
T0=var('T0')
eqd2=diff(T,t)==-k*(T-Tu)
sol2(k,T0,Tu)=desolve(eqd2,[T,t],ics=[0,T0])
sol2(k,T0,Tu)
```

```sage
plot([sol2(1,2,3),sol2(3,2,1),sol2(1,1,1)],t,-2,2)
```

```sage
x=function('x')(t)
r=var('r')
eqd3=diff(x,t)==r*x
sol3(r,x0)=desolve(eqd3,[x,t],ics=[0,x0])
sol3
```

```sage
plot([sol3(1,1),sol3(2,2)],t,-2,2)
```
