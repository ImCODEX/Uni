```python
x=var('x')
y=function('y')(x)
eqd=diff(y,x)==2*y
desolve(eqd,y)
```

```python
desolve(eqd,y,show_method=True)
```

```python
sol=desolve(eqd,y)
_C=var('_C')
sol.substitute(_C==1)
```

```python
sol1=sol.substitute(_C==1)
plot(sol1,x,-2,2)
```

```python
eqdb=diff(y,x)==y/x
desolve(eqdb,y)
```

```python
eqdc=diff(y,x)==2*x*(1+y*y)
desolve(eqdc,y)
```

```python
yy=var('yy')
_C=var('_C')
f(x,yy,_C)=desolve(eqdc,y).substitute(_C==0).substitute(y(x)==yy)
implicit_plot(f(x,yy,_C),(x,-5,5),(yy,-5,5))
```
